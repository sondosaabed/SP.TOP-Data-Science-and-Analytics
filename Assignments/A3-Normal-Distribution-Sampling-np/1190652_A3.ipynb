##### <i><b>1-2-3 Rule (empirical rule)</b></i>
It is a rule that relates the standard deviations to the area under the normal curve. The follwing are the rules: [3]
- One standard deviation on either side of the mean of a normal curve will encompass approximately 68% of the area under the curve.
- Two standard deviations on either side of the mean of a normal curve will encompass approximately 95% of the area under the curve.
- Three standard deviations above and below the mean will encompass slightly more than 99% of the area under the curve.

##### <i><b>Data Sampling</b></i> [8]
<p><b>A Sample is:</b> A subset of the population. It is used to generlize findings to the population. Usually because the population has large size that will be costing in terms of computations, time consuming and analyitics.</p>  
<p><b>Representative sample:</b> A sample in which the characteristics of the individuals closely match the characteristics of the overall population.</p>

<p>There are diffrent ways to obtain samples, the method could be simple random, systematic random, and Stratified random.</p>

There are diffrent factors to be considered when sampling:
- Population size: large sizes needs larger samples.
- Confidence level: The higher the confidence level, the larger your sample needs to be. Like: 90%, 95%, 99%.
- Margin of error to be tolerated such as +/- 5%.

Sampling error is always there because the sample will never be perfectly representative of the population.

<b>Random Sample:</b> means that Every unit or case in the population has an equal chance of being selected; the selection of any case or unit can in no way affect the selection of any other unit or case; the cases or units are selected in such a way that all combinations are possible.
<p><b>Standard error of the mean:</b> The standard error of the mean is the standard deviation of the sampling distribution of sample means. Standard error matters because it helps you estimate how well your sample data represents the whole population.</p>

### <b>Normal Distribution Generation</b>
- This is the first step of the experiment, Generation of a normally distributed population.

```python
def normal_rand_sample(avg = 1500, std = 700, size=10000000, seed =1190652):
    """
    This is a wrap function that calls the normal with diffrent params. values
    it also sets the seed of random locally to a specific seed.
    The deafult values are set as required in the experimnet
    Args:
        avg (double): the average of the normal distribution
        std (double): the standard deviation of the normal distribution
        size (integer): the size of the population
        seed (integer): the random seed
    Returns:
        (1D np.array) contains the generated sample with specific args
    """
    random.seed(seed)
    return random.normal(loc = avg, scale=std, size = size) 
```

- <b>Desriptive statics on an np array:</b>

```python
def get_descriptive_statics(data = None):
    """
    This function calucatles the descriptive statics of the data
    Args:
        data (1D np.array): data array
    Returns:
        array that contains descriptive statics
    """
    return [len(data), np.mean(data), np.std(data), np.max(data), np.min(data), np.median(data)]
```

```python
def print_descriptive_statics(statics = None):
    """
    This function prints teh statics of specifc np.array
    Args:
        statics (array): contains ddescriptive statics
    Returns:
        Nothings, prints the statics
    """
    print("Size of data is:", statics[0])
    print("The average of the data is:", statics[1])
    print("The standard deviation of the data is:", statics[2])
    print("The maximum value of the data is: ", statics[3])
    print("The minimum value of the data is: ", statics[4])
    print("The median in the data:", statics[5])
```

```python
def get_diff_sum(all_statics_samples, original):
    """
    This function calcuatles the errors or diffrences with the original data and the samples
    Args:
        all_statics_samples (1D array): statics arrays of diffrent samples
    Returns:
        diffrences_sums (1D array): contains the errors
    """
    diffrences_sums = []

    for static in all_statics_samples:
        diffrences_sums.append(
            np.sum(
                np.array(np.abs(original[1:])) - np.abs(np.array(static))))
        np.argmin(diffrences_sums)
        
    return diffrences_sums
```

```python
all_statics_samples = [statics_1[1:], statics_2[1:], statics_3[1:], statics_4[1:]] # because the first valuse is size, so exc;ude them
diffrences_sums = get_diff_sum(all_statics_samples, statics)
print("Diffrences sums for all samples:", diffrences_sums)
print("The minimum diffrences sum in samples is: ", np.min(diffrences_sums))
print("The index of the most representive sample is: ", np.argmin(diffrences_sums)) # since the first index was excluded (Zero indexing)
```

<b>3. By Standard Error of means:</b> let's say that the confidence level is 90%. Z is the score value (it is chosen due the abstarction of the data, meaning it's not critical to have high confidence)...
The std of the population is (σ) 700.045, the samples sizes (n) are [10, 100, 1000, 10000].
`SE = Z * σ/√n`

```python
def sem(z=1.645, std = 700.045, sample_size=0):
    """
    This function calculates the Standard error of mean.
    Args:
        z (double): confidence level is taken from the z-score table (90% -> 1.645)
        std (double): standard deviation of population
        sample_size (int): 
    Retruns:
        SEM value calcualted 
    """
    return z *(std / np.sqrt(sample_size))
```

```python
def get_sems(sample_sizes = None):
    """
    This function calcuates the sems of an array of sample sizes
    Args:
        sample_sizes (array): contains sizes of diffrent samples
    Return:
        (np.array): contains the SEM for each size
    """
    sems = []
    for size in sample_sizes:
        sems.append(sem(sample_size=size))
    return np.array(sems)
```

```python
sample_sizes = [10, 100, 1000, 10000]
sems = get_sems(sample_sizes)
print("Standard Error of means for diffrent samples: ", sems)
print("The minimum sample error is: ", np.min(sems))
print("The minmum sample error with inxed: ", np.argmin(sems))
```

```python
def get_quartile(data, quartile):
    """
    This is a wrap function that uses percentiles to calcualte the quartiles.
    Args:
        data (1D np.array): array of data
        quartile (int): the specific percentile
    Return:
        The value that the (quartile) percentage of the data is below it
    """
    return np.percentile(data, quartile)
```

```python
def plot_for_quartiles(data = None, quartiles =None):
    """
    [7]
    This function plots a box plot for quartiles visualization
    Args:
        data (1D np.array): data
        quartiles (array): contains the calcuated quartiles
    Returns:
        Nothing, plots the boxplot
    """
    plt.figure(figsize=(10, 5))
    plt.boxplot(data, vert=False, widths=0.5)
    colors =['red', 'blue', 'green']
    horizontalalignments = ['right', 'center', 'left']
    labels = ['25%', '50%', '75%']
    
    for q, (quartile, color, halign, label) in enumerate(zip(quartiles, colors, horizontalalignments, labels), start=1):
        plt.text(quartile, 1.02, f'{label} ({quartile:.2f})', horizontalalignment=halign, fontsize=9, color=color)

    plt.xlabel('Values')
    plt.title('Boxplot of Most represntaive Sample')
    plt.show()
```
