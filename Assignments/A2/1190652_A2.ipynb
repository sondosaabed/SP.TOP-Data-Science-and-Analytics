- To decide which grocery to chose, this is a minimization problem, where the best solution is the min of sum of the prices for the required list to buy.
- Such problem could be approached diffrently, but what I have in mind is the exhastive solution where I calculate the muliplications for all the possible combinations of groceries and quantities to find the minimum total cost.

##### Necessary imports
- First we import necessary libraries  

```python
import numpy as np
import os
```

- This is given:

```python
amounts = np.array([2, 1.5, 3, 6, 1, 1, 2, 3])
print("Amounts: ", amounts)
```

##### Load Files 
- First Let's read the files into numpy arrays, knowing that the groceries directory is in the same parent directory as this notebook.

```python
def get_np_arr(path):
    """
    This funtion is used to load npy files into np array object
    Args: 
        - path: Takes a path of a directory
    Returns:
        - Retunrs np array loaded from that file
    """
    return np.load(path)
```

- For each path in the grocieries directory call the `get_np_arr` funtion
- Use np array for fast multiplizations

```python
def get_names_prices(groceries_dir = "./groceries"):
    """
    This funtion loads the files into np.arrays
    Args:
        - groceries_dir: the deafult value is the relative path of the groceries
    Returns:
        - groceries_prices: 2D np.array with the prices
        - groceries_names: 1D np.array with the names of the groceraries
    """
    groceries_prices = []
    groceries_names = []

    for path in os.listdir(groceries_dir):
        groceries_names.append(path.strip(".npy"))
        groceries_prices.append(get_np_arr(groceries_dir + "/" + path))
    return np.array(groceries_prices), np.array(groceries_names)
```

- Looking at and printing the data:

```python
groceries_prices, groceries_names = get_names_prices()
print("Grocesseries Names: ", groceries_names)
print("Prices: ", groceries_prices)
```

##### Mulitplications for minimization
- At this point we have two np arrays, one contains the grociereis, the other contains the amounts.
    - `amounts` and `groceries_prices`

```python
def get_min_sum(groceries_prices=groceries_prices, amounts=amounts, groceries_names=groceries_names):
    """
    This function performs mulitplication with amounts and groceries_prices
    Args:
        - groceries_prices: 2D np.array of prices from the file
        - amounts: 1D np.array of the amounts required
    Return:
        - min: the min total price
        - total_prices.argmin(): the index to get the name of the grocearry
    """
    total_prices = []
    for grocerry in groceries_prices:
        mulitply = np.multiply(grocerry, amounts)
        print("Price with amounts: ", mulitply)
        
        total = np.sum(mulitply)
        print("Total price: ", total)

        total_prices.append(total)
 
    return np.array(total_prices)
```

- calling the function and getting the prices and the sums:

```python
total_prices = get_min_sum()
```

```python
print("All sum Prices: ", total_prices)
print("Min price: ", total_prices.min())
print("Index of the min price: ", total_prices.argmin())
```

- Using the min index to get the grocessary name:

```python
print("The groceary that I wil use is: ", groceries_names[total_prices.argmin()])
```
